```python
new_df = pd.read_csv('netflix_cleaned2.csv')
new_df.head(20)
```

```python
top10countries = new_df.groupby('country')['country'].count().sort_values(ascending= False).head(10).reset_index(name= 'count')
v = top10countries[top10countries['country'] != "Unknown"]
v 
```

```python
sns.barplot(data = v,
            x = 'count',
            y = 'country')

plt.title('TOP 10 COUNTRIES WITH THE MOST MOVIES')
plt.ylabel('Countries')
plt.show()
```

```python
genres = (
    new_df["listed_in"]
      .str.split(",")
      .explode()
      .str.strip()
      .value_counts()
      .head(10)
      .reset_index()
)

genres
```

```python
sns.barplot(data = genres,
            x = 'count',
            y = 'listed_in')

plt.title('TOP 10 FREQUENTED GENRES')
plt.xlabel('Frequency')
plt.ylabel('Genres')
plt.show()
```

```python
cast_member = (
    new_df["cast"]
      .str.split(",")
      .explode()
      .str.strip()
      .loc[lambda x : x != "Not Applicable"]
      .value_counts()
      .head(10)
      .reset_index()
)

cast_member
```

```python
ax=sns.barplot(data = cast_member,
            x = "count",
            y = "cast")

ax.bar_label(ax.containers[0])


plt.title('Top 10 Actors')
plt.xlabel('Number of Titles')
plt.ylabel('Actors')
plt.show()
```
